# Plotando uma imagem com OpenCV

 importando biblioteca do OpenCV

```python
import cv2
```

Importando matplotlib.pyplot para plotar a imagem

```python
import matplotlib.pyplot as plt
```

Lendo e plotando imagem 

```python
img = cv2.imread('./python.png')
plt.imshow(img)
```

Verificando o shape da imagem

```python
img.shape
```

OBS: Notem que o OpenCV carrega uma imagem RGB ao contrario, com sua cores invertidas. Devemos corrigir isso convertendo a imagem de BGR para RGB.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```
